```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
for ppppp in ['High','Open','Close','Low']:
    data = pd.read_csv('../src/csv/NSE-APTECHT.csv')
    data = data.drop(columns=['Last','Turnover (Lacs)'])
    # Invert Data
    data = data.iloc[::-1,:].reindex(index=data.index[::-1])
    data.tail()

    scl = MinMaxScaler()
    cl = data[ppppp].reshape(data[ppppp].shape[0],1)
    cl = scl.fit_transform(cl)
    print(cl)

    # Tuneable params : lb

    # length of batch = lb
    def splitData(data,lb):
        X, y = [],[]
        for i in range(len(data)-lb-1):
            X.append(data[i:i+lb,0])
            y.append(data[i+lb,0])
        return np.array(X), np.array(y)

    X,y = splitData(cl,7)
    X_train, X_test = X[:int(X.shape[0]*0.8)], X[int(X.shape[0]*0.8):]
    y_train, y_test = y[:int(y.shape[0]*0.8)], y[int(y.shape[0]*0.8):]
    print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

    # model

    model = Sequential()
    model.add(LSTM(256, input_shape=(7,1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

    # Reshaping Data for (Sample, Timestep, Features)
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    history = model.fit(X_train, y_train, epochs=75, validation_data=(X_test,y_test), shuffle=False)

    # plt.plot(history.history['loss'])
    # plt.plot(history.history['val_loss'])


    Xt = model.predict(X_test)
    plt.plot(scl.inverse_transform(y_test.reshape(-1,1)))
    plt.plot(scl.inverse_transform(Xt))


    act = []
    predicted = []
    i=120
    Xt = model.predict(X_test[i].reshape(1,7,1))
    print("predicted:{0}, actual{1}".format(scl.inverse_transform(Xt), scl.inverse_transform(y_test[i].reshape(-1,1))))
    predicted.append(scl.inverse_transform(Xt))
    act.append(scl.inverse_transform(y_test[i].reshape(-1,1)))

    model.save('../src/models/APTECHT_'+ppppp+'.h5')
```

```python
class Predict(object):
    
    lb = 7 
    
    def __init__(self, name, ticker, time):
        from keras.model import load_model 
        import pandas as pd
        from dateutil import parser
        from sklearn.preprocessing import MinMaxScaler

        self.NAME    = name
        self.TICKER  = ticker
        self.TIME    = time
        self.MODEL   = self.load_model()
        self.RESULTS = self.load_dictionary()
        self.DATA    = self.load_data()
        
    def load_dictionary(self):
        return {'Open' : 0,
                'Close': 0,
                'High' : 0,
                'Low'  : 0  }
    
    def load_data(self):
        table = pd.read_csv('NSE-'+self.TICKER+'.csv')
        table['Date'] = pd.to_datatime(table['Date'])
        table = table.iloc[::-1,:]
        return table
    
    def load_model(self):
        open_model  = load_model(self.TICKER+'_Open.h5')
        close_model = load_model(self.TICKER+'_Close.h5')
        high_model  = load_model(self.TICKER+'_High.h5')
        low_model   = load_model(self.TICKER+'_Low.h5')
        return {'Open' :open_model,
                'Close':close_model,
                'High' :high_model,
                'Low'  :low_model   }
    
    def predict_values(self):
        for key, value in self.RESULTS.items():
            scl = MinMaxScaler()
            data = self.DATA
            cl = self.DATA[key].reshape(self.DATA.shape[0],1)
            cl = scl.fit_transform(cl)
            indx = data.index[data.Time == self.TIME]
            test = data[indx:indx+self.lb,0]
            test = test.reshape(*test.shape,1)
            predicted_value = self.MODEL[key].predict(test)
            self.RESULTS[key] = predicted_value
        return self.RESULTS
```
